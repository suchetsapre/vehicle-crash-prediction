```python
#Set the location and name of the cfg file
cfg_file = './yolov3.cfg'

# Set the location and name of the pre-trained weights file
weight_file = './yolov3.weights'

# Set the location and name of the COCO object classes file
namesfile = './coco.names'

# Load the network architecture
m = Darknet(cfg_file)

# Load the pre-trained weights
m.load_weights(weight_file)

# Load the COCO object classes
class_names = load_class_names(namesfile)
```

```python
# We resize the image to the input width and height of the first layer of the network.    

vidcap = cv2.VideoCapture('./videos/training/positive/000003.mp4')
success, image = vidcap.read()
count = 0
frames = []
while success:
    frames.append(np.array(Image.fromarray(image)))
    #frames.append(np.array(Image.fromarray(image).convert('L')))
    #plt.imshow(image)
    #plt.show()
    #time.sleep(0.5)# 
    #plt.close('all')
    success,image = vidcap.read()
    count += 1
print(count)

original_image = frames[50]

resized_image = cv2.resize(original_image, (m.width, m.height))

# Display the images
plt.subplot(121)
plt.title('Original Image')
plt.imshow(frames[65])
```

```python

# Set the default figure size
#plt.rcParams['figure.figsize'] = [24.0, 14.0]

# Load the image
#img = cv2.imread('./images/dog.jpg')
img = frames[44]
# Convert the image to RGB
original_image = img#cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# We resize the image to the input width and height of the first layer of the network.    
resized_image = cv2.resize(original_image, (m.width, m.height))

# Set the IOU threshold. Default value is 0.4
iou_thresh = 0.4

# Set the NMS threshold. Default value is 0.6
nms_thresh = 0.6

# Detect objects in the image
boxes = detect_objects(m, resized_image, iou_thresh, nms_thresh)

# Print the objects found and the confidence level
#print_objects(boxes, class_names)

#Plot the image with bounding boxes and corresponding object class labels
plot_boxes(original_image, boxes, class_names, plot_labels = True)
```

```python
gray = resized_image.copy()
print(resized_image.shape)
for (x,y,w,h,a,b,c) in boxes:
    cv2.rectangle(gray,((x) * 416, (y) * 416),((x+w) * 416,(y+h) * 416) , (0, 255, 0), 3)
```

```python
plt.imshow(gray)
```

```python
# ensure at least one detection exists
if len(boxes) > 0:
    # loop over the indexes we are keeping
    for i in range(len(boxes)):
        # extract the bounding box coordinates
        (x, y) = (boxes[i][0], boxes[i][1])
        (w, h) = (boxes[i][2], boxes[i][3])
        cls_id = boxes[i][6]
        conf_interval = boxes[i][4]
        # draw a bounding box rectangle and label on the image
        color = (0, 255, 0)#[int(c) for c in COLORS[classIDs[i]]]
        cv2.rectangle(original_image, (x, y), (x + w, y + h), color, 2)
        text = "{}: {:.4f}".format(cls_id, conf_interval)
        cv2.putText(original_image, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX,
            0.5, color, 2)
 
# show the output image
cv2.imshow("Image", original_image)
cv2.waitKey(0)
```
